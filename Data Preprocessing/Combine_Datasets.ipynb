## **Combine Dataset**

````python
import pandas as pd

# Store DF

stores = pd.read_csv('/content/drive/MyDrive/GUVI FINAL PROJRCT/stores_data_set.csv')


# Sales DF

sales  = pd.read_csv('/content/drive/MyDrive/GUVI FINAL PROJRCT/sales_data_set.csv')

# Features DF

features = pd.read_csv('/content/drive/MyDrive/GUVI FINAL PROJRCT/Features_data_set.csv')
````

### **Combine sales And Feature Dataset**

````python
sales_1 = sales.loc[:210785]

sales_2 = sales.loc[210786:]
````

````python
sales_1
````

````python
sales_2
````

````python
sales_1.columns,features.columns
````

````python
df_1 = pd.merge(sales_1,features,on=['Store','Date','IsHoliday'],how='left')
````

````python
df_1
````

````python
df_2 = pd.merge(sales_2,features,on=['Store','IsHoliday','Date'],how='left')
````

````python
df_2
````

````python
df_1.head()
````

````python
df_2.head()
````

````python
df = pd.concat([df_1,df_2],ignore_index=True)
````

````python
df
````

````python
df_new = df.copy()
````

````python
df_new['date'] = pd.to_datetime(df_new['Date'],format='%d/%m/%Y')
````

````python
df_new.dtypes
````

### **Non presented dates From 2012 to 2013 years**

````python
missing_dates = features[~features['Date'].isin(df_new['Date'])].reset_index(drop=True)
````

````python
missing_dates
````

````python
missing_dates['date'] = pd.to_datetime(missing_dates['Date'],format='%d/%m/%Y')
````

````python
df_new_1 = df_new[['Store', 'Dept']]
````

````python
df_new_1 = df_new_1.drop_duplicates().reset_index(drop=True)
````

````python
df_new_1
````

````python
df_new_1.Store.nunique()
````

````python
df_new_1.Dept.nunique()
````

````python
df_new_2 = pd.merge(df_new_1,missing_dates,on=['Store'],how='outer')
````

````python
df_new_2
````

````python
df_new_2.drop('date',axis=1,inplace=True)
````

````python
df.shape,df_new_2.shape
````

````python
df.columns
````

````python
df_new_2.columns
````

````python
csv = pd.concat([df,df_new_2],ignore_index=True)

csv.shape
````

````python
csv
````

````python
csv['Date'] = pd.to_datetime(csv['Date'],format='%d/%m/%Y')

csv['Day'] = pd.to_datetime(csv['Date'],format='%d/%m/%Y').dt.day

csv['Month'] = pd.to_datetime(csv['Date'],format='%d/%m/%Y').dt.month

csv['Year'] = pd.to_datetime(csv['Date'],format='%d%m%/%Y').dt.year
````

````python
csv
````

### **Combine Final Csv and Store**

````python
stores
````

````python
df = pd.merge(csv,stores,on=['Store'],how='left')

df
````

````python
df['IsHoliday'] = df['IsHoliday'].map({False:0,True:1})

df['Type'] = df['Type'].map({'A':1,'B':2,'C':3})
````

````python
df.head()
````

````python
df = df.sort_values(by=['Date','Store']).reset_index(drop=True)  # Sort df if u want
````

````python
df.to_csv('Combined_Datasets.csv',index=False)
````
